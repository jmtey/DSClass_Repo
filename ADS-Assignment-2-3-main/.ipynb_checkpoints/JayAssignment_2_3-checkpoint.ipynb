```python
#import the required libraries

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
#import the dataset

att_df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
att_df.head(10)
```

```python
#Understanding the dataset --checking if any column has missing values, data types, how many columns and rows?
att_df.info()
```

```python
#Understanding some basic stat
att_df.describe()
```

```python
#check the overall attrition --actual numbers
att_count=att_df.Attrition.value_counts()
att_count
```

```python
#check the overall attrition --percentages
att_count_per=att_df.Attrition.value_counts(normalize=True)*100
att_count_per 
```

```python
pd.crosstab(att_df['DistanceFromHome'],att_df['Attrition'])
```

```python
pd.crosstab(att_df['JobRole'],att_df['Attrition'])
```

```python
# Conclusion:Employees attrition is higher on staff working close to the office i.e distance of less than 10

plt.figure(figsize = (15,8))
sns.displot( x= "DistanceFromHome", hue = "Attrition", data =att_df, kde = "True")
plt.title("Attrition Vs Distance from Home to the office")
plt.xlabel("Distance From Home")
```

```python
#Conclusion:Three job roles of Labaratory Technicians,Research Scientist and Sales Executives contributes the most number of staff who have left

plt.figure(figsize = (14,8))
sns.countplot( y= "JobRole", hue = "Attrition", data =att_df)
plt.title("Attrition vs Job Roles")
plt.ylabel("Job Roles")
```

```python
#Conclusion: Attrition is higher among emloyees with low salary i.e below 5000

sns.displot( x= "MonthlyIncome", hue = "Attrition", data =att_df, kde = "True")
plt.title("Attrition Vs MonthlyIncome")
plt.xlabel("MonthlyIncome")
```

```python
#Conclusion: Employees with the Education of 3 do leave the company often as opposed to others

sns.displot( x= "Education", hue = "Attrition", data =att_df)
plt.title("Attrition Vs Education")
plt.xlabel("Education")
```
